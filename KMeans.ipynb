```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/home/admin1/Downloads/iris.csv')
```

```python
df
```

```python
features=df.drop('Species',axis=1)
```

```python
features
```

```python
scaler =StandardScaler()
data_scaled=scaler.fit_transform(features)
```

```python
data_scaled
```

```python
wcss=[]


for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)
```

```python
print(wcss)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,11),wcss, marker='o',color='b')
plt.title('Elbow Method for optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('wcss')
plt.grid(True)
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
ymeans=kmeans.fit_predict(data_scaled)
```

```python
plt.scatter(data_scaled[ymeans==0,0],data_scaled[ymeans==0,1],s=100,c='purple',label='Iris-setosa')
plt.scatter(data_scaled[ymeans==1,0],data_scaled[ymeans==1,1],s=100,c='orange',label='Iris-versicolor')
plt.scatter(data_scaled[ymeans==2,0],data_scaled[ymeans==2,1],s=100,c='green',label='Iris-virginica')
```
